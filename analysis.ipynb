```python
import easystac as es
import json
from  shapely.geometry import shape
```

```python
with open('remote_sensing_challenge_AOI.geojson') as file:
    geo_file =  json.load(file)
    features = geo_file['features']
    for feat in features:
        geom = shape(feat['geometry'])
        E84_S2_L2A = (es.ImageCollection('sentinel-s2-l2a-cogs')
                    .fromSTAC('https://earth-search.aws.element84.com/v0')
                    .filterBounds(geom)
                    .filterDate("2020-01-01","2022-12-01")
                    .getInfo(epsg = 4326, resolution = 20))
E84_S2_L2A
```
